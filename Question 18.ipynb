3
7 4
2 4 6
8 5 9 3

That is, 3 + 7 + 4 + 9 = 23.

Find the maximum total from top to bottom of the triangle below:

75

NOTE: As there are only 16384 routes, it is possible to solve this problem by trying every route. However, Problem 67, is the same challenge with a triangle containing one-hundred rows; it cannot be solved by brute force, and requires a clever method! ;o)

```python
triangle_raw ="""75
95 64
17 47 82
18 35 87 10
20 04 82 47 65
19 01 23 75 03 34
88 02 77 73 07 63 67
99 65 04 28 06 16 70 92
41 41 26 56 83 40 80 70 33
41 48 72 33 47 32 37 16 94 29
53 71 44 65 25 43 91 52 97 51 14
70 11 33 28 77 73 17 78 39 68 17 57
91 71 52 38 17 14 91 43 58 50 27 29 48
63 66 04 68 89 53 67 30 73 16 69 87 40 31
04 62 98 27 23 09 70 98 73 93 38 53 60 04 23"""

triangle_split = triangle_raw.split()
triangle_length = 15
zero = 0
count = 0

triangle_grid = [[0 for row in range(triangle_length)] for col in range(triangle_length)]

for row in range(triangle_length):
    for col in range(14-row):
        del triangle_grid[row][14-col]
    for col in range(row + 1):
        triangle_grid[row][col] = int(triangle_split[count])
        count += 1

for row in range(triangle_length):
    for col in range(len(triangle_grid[row])):
        print("{:02d}".format(triangle_grid[row][col]), end=' ')
    print()
```

```python
while len(triangle_grid) >= 1:
    row = len(triangle_grid) - 1
    col_length = len(triangle_grid[row]) - 1

    for col in range(col_length):
        options = [triangle_grid[row][col],triangle_grid[row][col+1]]
        largest_num = max(options)
        triangle_grid[row-1][col] += largest_num

    del triangle_grid[row]

    for row_temp in range(len(triangle_grid)):
        for col_temp in range(len(triangle_grid[row_temp])):
            print("{:03d}".format(triangle_grid[row_temp][col_temp]), end=' ')
        print()
```
